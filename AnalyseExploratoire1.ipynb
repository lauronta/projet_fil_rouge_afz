```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import ElasticNet
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import max_error
from sklearn.metrics import r2_score
```

```python
Y_t["UFV"]
```

```python
target = ["UFV", "UFL", "PDI", "BPR"]
models = [LinearRegression(), Ridge(), SVR(), DecisionTreeRegressor(), 
          RandomForestRegressor(), GradientBoostingRegressor(), 
          XGBRegressor(), MLPRegressor(), KNeighborsRegressor()]

mape_tot = []
max_error_tot = []
mae_tot = []
R2_tot = []

for tar in target:
    
    Y_t_temp = Y_t[tar]
    
    X_train, X_test, Y_train, Y_test = train_test_split(X_t, Y_t_temp, test_size=0.33, random_state=0)
    
    mape_temp = []
    max_error_temp = []
    mae_temp = []
    R2_temp = []
    
    for mod in models :
        
        model_used = mod
        model_used.fit(X_train, Y_train)
        Yhat = model_used.predict(X_test)
        
        mape = mean_absolute_percentage_error(Y_test, Yhat).round(4)
        mape_temp.append(mape)
        
        error_max = max_error(Y_test, Yhat).round(4)
        max_error_temp.append(error_max)
        
        mae = mean_absolute_error(Y_test, Yhat).round(4)
        mae_temp.append(mae)
        
        r2 = r2_score(Y_test, Yhat)
        R2_temp.append(r2)
        
    
    mape_tot.append(mape_temp)
    max_error_tot.append(max_error_temp)
    mae_tot.append(mae_temp)
    R2_tot.append(R2_temp)
    
model_names = [mod.__class__.__name__ for mod in models] 

mape_df = pd.DataFrame(mape_tot, columns=model_names, index=target)
max_error_df = pd.DataFrame(max_error_tot, columns=model_names, index=target)
mae_tot_df = pd.DataFrame(mae_tot, columns=model_names, index=target)
R2_tot_df = pd.DataFrame(R2_tot, columns=model_names, index=target)

mape_df = mape_df.T
max_error_df = max_error_df.T
mae_tot_df = mae_tot_df.T
R2_tot = R2_tot_df.T
```
